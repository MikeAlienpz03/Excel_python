**Gráficos y subgráficos**

* Los gráficos en matplotlib residen dentro de un objeto Figura. Puede crear una nueva figura con plt.figure:

* La función add_subplot en Matplotlib se utiliza para agregar subgráficos (también conocidos como subplots) a una figura. Los subplots se organizan en una cuadrícula, y puedes especificar en qué posición colocar cada subplot.

* ax = fig.add_subplot(nrows, ncols, index)



```python
fig = plt.figure()
# Añadir subplots a la figura
ax1 = fig.add_subplot(2, 2, 1)  # Primer subplot en una cuadrícula de 2x2
ax2 = fig.add_subplot(2, 2, 2)  # Segundo subplot en una cuadrícula de 2x2
ax3 = fig.add_subplot(2, 2, 3)  # Tercer subplot en una cuadrícula de 2x2
ax4 = fig.add_subplot(2, 2, 4)


# Opcional: Añadir gráficos a los subplots
ax1.plot([1, 2, 3], [1, 4, 9])
ax2.plot([1, 2, 3], [1, 2, 3])
ax3.plot([1, 2, 3], [1, 0, 1])

# Mostrar la figura con los subplots
plt.show()
```

**Notas**

* sharex=True: Este argumento indica que todos los subplots compartirán el mismo eje x. Esto significa que los límites y las etiquetas del eje x serán los mismos para todos los subplots.

* sharey=True: Este argumento indica que todos los subplots compartirán el mismo eje y. Esto significa que los límites y las etiquetas del eje y serán los mismos para todos los subplots.

```python
fig, axes = plt.subplots(2,2,sharex=True,sharey=True)
for w in range(2):
    for j in range(2):
        axes[w,j].hist(np.random.randn(500),bins=50,color="r",alpha=0.5)
    plt.subplots_adjust(wspace=0,hspace=0)
```

**Ticks, Labels, and Legends**

* La interfaz de pyplot, diseñada para un uso interactivo, consta de métodos como xlim, xticks y xticklabels. Éstos controlan el rango del gráfico, la ubicación de las marcas y las etiquetas de las marcas, respectivamente. Se pueden utilizar de dos maneras:

* Llamada sin argumentos devuelve el valor actual del parámetro. Por ejemplo  plt.xlim() devuelve el rango de trazado actual del eje X

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(data, 'k-', drawstyle='steps-post', label='steps-post')
# Establecer los límites del eje x con subplot
ax.set_xlim([0,40])
ax.legend(loc='best')
```

* Para cambiar los ticks del eje X, lo más fácil es usar set_xticks y set_xticklabels. El primero indica a matplotlib dónde colocar las marcas a lo largo del intervalo de datos; por defecto, estas ubicaciones serán también las etiquetas. Pero podemos establecer cualquier otro valor como etiqueta utilizando set_xticklabels:

* **ax.set_xticks([0, 250, 500, 750, 1000])** establece las ubicaciones de las marcas en el eje x en los puntos especificados: 0, 250, 500, 750 y 1000. en el grafico

* **ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=30, fontsize='small')** establece las etiquetas de las marcas en el eje x como 'one', 'two', 'three', 'four' y 'five'. Además, rotation=30 rota las etiquetas 30 grados en sentido antihorario y fontsize='small' establece el tamaño de la fuente de las etiquetas como pequeño.

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())
ax.set_title('My first matplotlib plot')
ax.set_xlabel('Stages')

ticks = ax.set_xticks([0, 250, 500, 750, 1000])
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'],rotation=30, fontsize='small')
```

**ax.annotate() para añadir anotaciones:**

* ax.annotate(label, xy=(date, spx.asof(date) + 50), xytext=(date, spx.asof(date) + 200), ...) agrega una anotación al gráfico en la posición especificada por xy.

* xy=(date, spx.asof(date) + 50) especifica la posición donde se dibujará la flecha de la anotación, desplazada verticalmente por 50 unidades desde el valor correspondiente en spx en la fecha date.

* xytext=(date, spx.asof(date) + 200) especifica la posición del texto de la anotación, desplazada verticalmente por 200 unidades desde el valor correspondiente en spx en la fecha date.

* arrowprops=dict(facecolor='black') define las propiedades de la flecha que conecta la anotación al punto especificado en xy.
horizontalalignment='left', verticalalignment='top' alinea el texto de la anotación de modo que esté a la izquierda y arriba de la posición de la flecha.

```python
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime

# Datos de ejemplo para ilustrar
dates = pd.date_range('2023-01-01', periods=5)
print(dates)
spx = pd.Series([1400, 1375, 1350, 1325, 1300], index=dates)

# Crear una figura y un subplot
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# Graficar la serie temporal SPX en el subplot
spx.plot(ax=ax, style='k-')

# Definir una fecha específica y su valor correspondiente en spx
date = datetime(2023, 1, 3)

# Agregar una anotación al gráfico en la posición especificada por xy
ax.annotate('Ejemplo de anotación', xy=(date, spx.asof(date) + 30),
            arrowprops=dict(facecolor='black'))

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime

# Crear una figura y un subplot
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# Leer los datos del archivo CSV (asegúrate de tener el archivo correcto y la ubicación adecuada)
data = pd.read_csv("../Files/spx.csv", index_col=0, parse_dates=True)

# Seleccionar la serie de datos SPX del DataFrame cargado
spx = data["SPX"]

# Graficar la serie SPX en el subplot
spx.plot(ax=ax, style='k-')

# Definir datos de crisis para anotaciones
crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]

# Añadir anotaciones al gráfico
for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 50),  # Posición de la anotación
                xytext=(date, spx.asof(date) + 200),    # Posición del texto de la anotación
                arrowprops=dict(facecolor='blue'),     # Propiedades de la flecha de la anotación
                horizontalalignment='left',             # Alineación horizontal del texto
                verticalalignment='top')                # Alineación vertical del texto

# Establecer los límites del eje x y y
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])

# Establecer el título del gráfico
ax.set_title('Important dates in 2008-2009 financial crisis')

# Mostrar el gráfico
plt.show()
```

**La función pd.crosstab()**

* en Pandas se utiliza para calcular una tabla de frecuencias cruzadas entre dos o más factores. Básicamente, te permite contar las frecuencias de ocurrencia de diferentes combinaciones de variables en forma de tabla.

* Consejos para usar pd.crosstab():
Variables categóricas: Idealmente, las variables que uses con pd.crosstab() deben ser categóricas o tener un número limitado de valores únicos para que la tabla de frecuencias cruzadas tenga sentido y sea manejable.

* Interpretación: La tabla de frecuencias cruzadas te permite ver cómo se distribuyen las observaciones a lo largo de múltiples dimensiones categóricas, lo cual es útil para entender la relación entre diferentes variables en tus datos.

* Flexibilidad: Puedes especificar múltiples columnas o índices para crear tablas de frecuencias cruzadas más complejas si tu conjunto de datos tiene más de dos dimensiones que te interesen analizar.

```python
tips = pd.read_csv("../Files/tips.csv")
print(tips.columns.tolist())
```

```python
party_counts = pd.crosstab(tips.day, tips.size)
party_counts
```

**El gráfico KDE (Kernel Density Estimation)**

* es una forma de visualización que se utiliza comúnmente en estadísticas y análisis de datos para estimar la distribución de probabilidad de un conjunto de datos

* Un tipo de gráfico relacionado es el gráfico de densidad, que se forma calculando una estimación de una distribución de probabilidad continua que podría haber generado los datos observados.

* Así, los gráficos de densidad
también conocidos como gráficos KDE (kernel density estimate). Usando plot con kind='kde' se obtiene un gráfico de densidad usando la mezcla estándar de normales KDE

**Uso en la práctica:**
El KDE es útil en diversas aplicaciones, como:

* Análisis exploratorio de datos: Para entender la forma y la dispersión de los datos.

* Comparación de distribuciones: Para visualizar cómo se superponen o difieren las distribuciones de diferentes conjuntos de datos.

* Modelado estadístico: Como parte de técnicas más avanzadas como la estimación de densidad no paramétrica.

```python
tips["tip_pct"].plot(kind="kde")
```

* **np.random.normal(media,desviacion_estandar,size)**

* Nos permite generar valores aleatorios distribuidos normalmente a partir de una media y desviacion estandar especificadas

```python
# Generar 200 números aleatorios distribuidos normalmente
# media = 0, desviacion estandar = 1
comp1 = np.random.normal(0, 1, size=200)

# Genera 200 números aleatorios con media 10 y desviación estándar 2
comp2 = np.random.normal(10, 2, size=200)

values = pd.Series(np.concatenate([comp1,comp2]))

#density=True 
# indica que se quiere mostrar densidades de probabilidad en lugar de frecuencias absolutas.
values.hist(bins=100,alpha=0.3,color="r",density=True)
values.plot(kind="kde",style="k--",linewidth=2)
plt.title('Histograma y KDE de valores')
plt.grid(True)
plt.legend(['KDE', 'Histograma'])
plt.ylabel('Densidad de probabilidad')
plt.xlabel('Valores')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
np.random.seed(0)
x = np.linspace(0, 10, 100)
y = np.sin(x) + np.random.normal(0, 0.1, 100)  # Función sinusoidal con ruido

# Calcular la media y la desviación estándar
mean = np.mean(y)
std_dev = np.std(y)

# Configurar el gráfico
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Datos')
plt.plot(x, np.ones_like(x) * mean, color='red', linestyle='dashed', label=f'Media: {mean:.2f}')
plt.fill_between(x, mean - std_dev, mean + std_dev, color='gray', alpha=0.2, label='± Desv. Est.')

# Mostrar leyenda y etiquetas
plt.legend()
plt.title('Gráfico de Líneas con Desviación Estándar')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)

# Mostrar gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=100)  # Generar datos normales

# Calcular la media y la desviación estándar
mean = np.mean(data)
std_dev = np.std(data)

# Configurar el gráfico
plt.figure(figsize=(8, 6))
plt.hist(data, bins=20, edgecolor='black', alpha=0.7)

# Añadir la media y la desviación estándar como líneas verticales
plt.axvline(mean, color='red', linestyle='dashed', linewidth=1.5, label=f'Media: {mean:.2f}')
plt.axvline(mean + std_dev, color='green', linestyle='dashed', linewidth=1.5, label=f'+1 Desv. Est.: {mean + std_dev:.2f}')
plt.axvline(mean - std_dev, color='green', linestyle='dashed', linewidth=1.5, label=f'-1 Desv. Est.: {mean - std_dev:.2f}')

# Mostrar leyenda y etiquetas
plt.legend()
plt.title('Histograma con Desviación Estándar')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.grid(True)

# Mostrar gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generar datos de una distribución normal
np.random.seed(0)
data = np.random.normal(loc=170, scale=10, size=1000)  # Media=170, Desviación estándar=10, 1000 datos

# Calcular percentil 25
percentil_25 = np.percentile(data, 25)

# Configurar el gráfico
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, color='blue', stat='density', bins=30)  # Gráfico de densidad
plt.axvline(percentil_25, color='red', linestyle='dashed', linewidth=2, label=f'Percentil 25: {percentil_25:.2f}')
plt.title('Distribución Normal con Percentil 25')
plt.xlabel('Alturas (cm)')
plt.ylabel('Densidad de Probabilidad')
plt.legend()

# Mostrar gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generar datos de una distribución normal
np.random.seed(0)
data = np.random.normal(loc=170, scale=10, size=1000)  # Media=170, Desviación estándar=10, 1000 datos

# Calcular percentiles 25 y 75
percentil_25 = np.percentile(data, 25)
percentil_75 = np.percentile(data, 75)

# Configurar el gráfico
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=True, color='blue', stat='density', bins=30)  # Gráfico de densidad
plt.axvline(percentil_25, color='red', linestyle='dashed', linewidth=2, label=f'Percentil 25: {percentil_25:.2f}')
plt.axvline(percentil_75, color='green', linestyle='dashed', linewidth=2, label=f'Percentil 75: {percentil_75:.2f}')
plt.title('Distribución Normal con Percentiles 25 y 75')
plt.xlabel('Alturas (cm)')
plt.ylabel('Densidad de Probabilidad')
plt.legend()

# Mostrar gráfico
plt.show()
```
