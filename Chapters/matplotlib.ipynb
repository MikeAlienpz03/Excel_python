**Dos Interfaces por el Precio de Una**

* Una característica potencialmente confusa de Matplotlib es su doble interfaz: una interfaz conveniente basada en el estado al estilo de MATLAB y una interfaz orientada a objetos más poderosa. Aquí destacaremos rápidamente las diferencias entre las dos.

```python
plt.figure()
plt.subplot(2,1,1) #rows, columns, panel number
plt.plot(x, np.sin(x))

#Create the second panel
plt.subplot(2,1,2) #rows, columns, panel number
plt.plot(x, np.cos(x))
```

**Interfaz Orientada a Objetos**

* La interfaz orientada a objetos está disponible para situaciones más complicadas y cuando deseas tener un control más preciso sobre tu figura. En lugar de depender de alguna noción de una figura o ejes "activos", en la interfaz orientada a objetos las funciones de trazado son métodos de objetos explícitos Figure y Axes. Para recrear el gráfico anterior utilizando este estilo de trazado, podrías hacer lo siguiente:

```python
fig, ax = plt.subplots(2)
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

**Ajustando el Gráfico: Colores y Estilos de Línea**

* El primer ajuste que es posible que desees hacer en un gráfico es controlar los colores y estilos de las líneas. La función plt.plot() acepta argumentos adicionales que se pueden usar para especificar estos detalles. Para ajustar el color, puedes usar la palabra clave color, la cual acepta una cadena que representa virtualmente cualquier color imaginable. El color se puede especificar de diversas maneras

```python
plt.style.use('seaborn-v0_8-talk')
plt.plot(x, np.sin(x - 0), color='blue')        
plt.plot(x, np.sin(x - 1), color='g')           
plt.plot(x, np.sin(x - 2), color='0.75')        
plt.plot(x, np.sin(x - 3), color='#FFDD44')     
# specify color by name
 # short color code (rgbcmyk)
 # Grayscale between 0 and 1
 # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 and 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python

plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');
# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

**Ajustando el Gráfico: Límites de los Ejes**

* Matplotlib hace un trabajo decente al elegir límites predeterminados para los ejes de tu gráfico, pero a veces es útil tener un control más fino. 

* La forma más básica de ajustar los límites de los ejes es utilizando los métodos **plt.xlim() y plt.ylim()**

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x, np.sin(x))
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

**Otro ejemplo**

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x, np.sin(x))
plt.xlim(10,0)
plt.ylim(1.2,-1.2)
```

* Un método útil relacionado es **plt.axis()** (nota la posible confusión entre "axes" con e y "axis" con i). 

* El método plt.axis() te permite establecer los límites tanto en el eje x como en el eje y con una sola llamada, pasando una lista que especifica **[xmin, xmax, ymin, ymax].**


```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x, np.sin(x))
plt.axis([-1,11,-1.5,1.5])
```

El método plt.axis() va incluso más allá, permitiéndote hacer cosas como ajustar automáticamente los límites alrededor del gráfico actual.

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x, np.sin(x))
ax.axis('tight')
```

* También permite especificaciones de nivel superior, como asegurar una relación de aspecto igual, de modo que en tu pantalla, una unidad en el eje x sea igual a una unidad en el eje y.

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x, np.cos(x))
ax.axis('equal')
```

**Etiquetado de Gráficos**

* Como última parte de esta sección, veremos brevemente cómo etiquetar los gráficos: títulos, etiquetas de ejes y leyendas simples.

* Los títulos y las etiquetas de los ejes son las etiquetas más simples. Hay métodos que se pueden utilizar para establecerlos rápidamente

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x, np.sin(x))
ax.set_title('a Sine Curve')
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')
```

* Cuando se muestran varias líneas dentro de un mismo eje, puede ser útil crear una leyenda del gráfico que etiquete cada tipo de línea. Nuevamente, Matplotlib tiene una manera incorporada de hacer esto rápidamente. **Se hace a través del método plt.legend().**

* Aunque hay varias formas válidas de usar esto, encuentro que es más fácil especificar la etiqueta de cada línea usando la palabra clave label de la función de trazado.


```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x, np.sin(x), '-g',label='sin(x)')
ax.plot(x, np.cos(x), ':g',label='cos(x)')
ax.axis('equal')
ax.legend(loc='best');
```

**Errores Comunes en Matplotlib**

* Si bien la mayoría de las funciones de plt se traducen directamente a métodos de **ax (como plt.plot() → ax.plot()**, **plt.legend() → ax.legend(), etc.)**, este no es el caso para todos los comandos. En particular, las funciones para establecer límites, etiquetas y títulos están ligeramente modificadas.

* plt.xlabel() → ax.set_xlabel()
* plt.ylabel() → ax.set_ylabel()
* plt.xlim() → ax.set_xlim()
* plt.ylim() → ax.set_ylim()
* plt.title() → ax.set_title()

* En la interfaz orientada a objetos para trazado, en lugar de llamar a estas funciones individualmente, a menudo es más conveniente utilizar el método ax.set() para establecer todas estas propiedades a la vez.

```python
fig = plt.figure()
# Agregar un conjunto de ejes a la figura
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0,10), ylim=(-2,2),
       xlabel='x', ylabel='sin(x)',
       title='A simple Plot')
```

**Diagramas de Dispersión Simples**

* Otro tipo de gráfico comúnmente utilizado es el diagrama de dispersión simple, un pariente cercano del gráfico de línea. En lugar de que los puntos estén unidos por segmentos de línea, aquí los puntos se representan individualmente con un punto, círculo u otra forma. Comenzaremos configurando el cuaderno para el trazado e importando las funciones que utilizaremos:

**Resumen**

* plt.axes(): Es más simple y directo, útil para crear gráficos básicos con un solo conjunto de ejes.

* fig.add_subplot(): Es más flexible y potente, especialmente cuando necesitas crear gráficos con múltiples subplots en la misma figura.

```python
x = np.linspace(0,10,30)
y = np.sin(x)

fig = plt.figure()
ax = plt.axes()
ax = plt.plot(x,y,'o',color='black')
```

**El tercer argumento en la llamada a la función es un carácter que representa el tipo de símbolo utilizado para el trazado.** 

* Así como puedes especificar opciones como '-' y '--' para controlar el estilo de la línea, el estilo del marcador tiene su propio conjunto de códigos de cadena cortos. La lista completa de símbolos disponibles se puede ver en la documentación de plt.plot, o en la documentación en línea de Matplotlib. La mayoría de las posibilidades son bastante intuitivas, y mostraremos algunos de los más comunes aquí.

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
    
plt.legend(numpoints=1)
plt.xlim(0, 1.8);
```

```python
fig = plt.figure()
ax = plt.axes()

ax.plot(x,y,'-ok');
```

Argumentos clave adicionales para plt.plot especifican una amplia gama de propiedades de las líneas y los marcadores.

```python
fig = plt.figure()
ax  = plt.axes()

ax.plot(x,y,'-p',color='gray',
        markersize=15,linewidth=4,
        markerfacecolor = 'white',
        markeredgecolor = 'gray',
        markeredgewidth = 2)
ax.set_ylim(-1.2,1.2);
```
