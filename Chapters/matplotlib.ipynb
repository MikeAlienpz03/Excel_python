**Dos Interfaces por el Precio de Una**

* Una característica potencialmente confusa de Matplotlib es su doble interfaz: una interfaz conveniente basada en el estado al estilo de MATLAB y una interfaz orientada a objetos más poderosa. Aquí destacaremos rápidamente las diferencias entre las dos.

```python
plt.figure()
plt.subplot(2,1,1) #rows, columns, panel number
plt.plot(x, np.sin(x))

#Create the second panel
plt.subplot(2,1,2) #rows, columns, panel number
plt.plot(x, np.cos(x))
```

**Interfaz Orientada a Objetos**

* La interfaz orientada a objetos está disponible para situaciones más complicadas y cuando deseas tener un control más preciso sobre tu figura. En lugar de depender de alguna noción de una figura o ejes "activos", en la interfaz orientada a objetos las funciones de trazado son métodos de objetos explícitos Figure y Axes. Para recrear el gráfico anterior utilizando este estilo de trazado, podrías hacer lo siguiente:

```python
fig, ax = plt.subplots(2)
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

**Ajustando el Gráfico: Colores y Estilos de Línea**

* El primer ajuste que es posible que desees hacer en un gráfico es controlar los colores y estilos de las líneas. La función plt.plot() acepta argumentos adicionales que se pueden usar para especificar estos detalles. Para ajustar el color, puedes usar la palabra clave color, la cual acepta una cadena que representa virtualmente cualquier color imaginable. El color se puede especificar de diversas maneras

```python
plt.style.use('seaborn-v0_8-talk')
plt.plot(x, np.sin(x - 0), color='blue')        
plt.plot(x, np.sin(x - 1), color='g')           
plt.plot(x, np.sin(x - 2), color='0.75')        
plt.plot(x, np.sin(x - 3), color='#FFDD44')     
# specify color by name
 # short color code (rgbcmyk)
 # Grayscale between 0 and 1
 # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 and 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python

plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');
# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

**Ajustando el Gráfico: Límites de los Ejes**

* Matplotlib hace un trabajo decente al elegir límites predeterminados para los ejes de tu gráfico, pero a veces es útil tener un control más fino. 

* La forma más básica de ajustar los límites de los ejes es utilizando los métodos **plt.xlim() y plt.ylim()**

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x, np.sin(x))
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

**Otro ejemplo**

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x, np.sin(x))
plt.xlim(10,0)
plt.ylim(1.2,-1.2)
```

* Un método útil relacionado es **plt.axis()** (nota la posible confusión entre "axes" con e y "axis" con i). 

* El método plt.axis() te permite establecer los límites tanto en el eje x como en el eje y con una sola llamada, pasando una lista que especifica **[xmin, xmax, ymin, ymax].**


```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x, np.sin(x))
plt.axis([-1,11,-1.5,1.5])
```

El método plt.axis() va incluso más allá, permitiéndote hacer cosas como ajustar automáticamente los límites alrededor del gráfico actual.

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x, np.sin(x))
ax.axis('tight')
```

* También permite especificaciones de nivel superior, como asegurar una relación de aspecto igual, de modo que en tu pantalla, una unidad en el eje x sea igual a una unidad en el eje y.

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x, np.cos(x))
ax.axis('equal')
```

**Etiquetado de Gráficos**

* Como última parte de esta sección, veremos brevemente cómo etiquetar los gráficos: títulos, etiquetas de ejes y leyendas simples.

* Los títulos y las etiquetas de los ejes son las etiquetas más simples. Hay métodos que se pueden utilizar para establecerlos rápidamente

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x, np.sin(x))
ax.set_title('a Sine Curve')
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')
```

* Cuando se muestran varias líneas dentro de un mismo eje, puede ser útil crear una leyenda del gráfico que etiquete cada tipo de línea. Nuevamente, Matplotlib tiene una manera incorporada de hacer esto rápidamente. **Se hace a través del método plt.legend().**

* Aunque hay varias formas válidas de usar esto, encuentro que es más fácil especificar la etiqueta de cada línea usando la palabra clave label de la función de trazado.


```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(x, np.sin(x), '-g',label='sin(x)')
ax.plot(x, np.cos(x), ':g',label='cos(x)')
ax.axis('equal')
ax.legend(loc='best');
```

**Errores Comunes en Matplotlib**

* Si bien la mayoría de las funciones de plt se traducen directamente a métodos de **ax (como plt.plot() → ax.plot()**, **plt.legend() → ax.legend(), etc.)**, este no es el caso para todos los comandos. En particular, las funciones para establecer límites, etiquetas y títulos están ligeramente modificadas.

* plt.xlabel() → ax.set_xlabel()
* plt.ylabel() → ax.set_ylabel()
* plt.xlim() → ax.set_xlim()
* plt.ylim() → ax.set_ylim()
* plt.title() → ax.set_title()

* En la interfaz orientada a objetos para trazado, en lugar de llamar a estas funciones individualmente, a menudo es más conveniente utilizar el método ax.set() para establecer todas estas propiedades a la vez.

```python
fig = plt.figure()
# Agregar un conjunto de ejes a la figura
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0,10), ylim=(-2,2),
       xlabel='x', ylabel='sin(x)',
       title='A simple Plot')
```

**Diagramas de Dispersión Simples**

* Otro tipo de gráfico comúnmente utilizado es el diagrama de dispersión simple, un pariente cercano del gráfico de línea. En lugar de que los puntos estén unidos por segmentos de línea, aquí los puntos se representan individualmente con un punto, círculo u otra forma. Comenzaremos configurando el cuaderno para el trazado e importando las funciones que utilizaremos:

**Resumen**

* plt.axes(): Es más simple y directo, útil para crear gráficos básicos con un solo conjunto de ejes.

* fig.add_subplot(): Es más flexible y potente, especialmente cuando necesitas crear gráficos con múltiples subplots en la misma figura.

```python
x = np.linspace(0,10,30)
y = np.sin(x)

fig = plt.figure()
ax = plt.axes()
ax = plt.plot(x,y,'o',color='black')
```

**El tercer argumento en la llamada a la función es un carácter que representa el tipo de símbolo utilizado para el trazado.** 

* Así como puedes especificar opciones como '-' y '--' para controlar el estilo de la línea, el estilo del marcador tiene su propio conjunto de códigos de cadena cortos. La lista completa de símbolos disponibles se puede ver en la documentación de plt.plot, o en la documentación en línea de Matplotlib. La mayoría de las posibilidades son bastante intuitivas, y mostraremos algunos de los más comunes aquí.

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
    
plt.legend(numpoints=1)
plt.xlim(0, 1.8);
```

```python
fig = plt.figure()
ax = plt.axes()

ax.plot(x,y,'-ok');
```

Argumentos clave adicionales para plt.plot especifican una amplia gama de propiedades de las líneas y los marcadores.

```python
fig = plt.figure()
ax  = plt.axes()

ax.plot(x,y,'-p',color='gray',
        markersize=15,linewidth=4,
        markerfacecolor = 'white',
        markeredgecolor = 'gray',
        markeredgewidth = 2)
ax.set_ylim(-1.2,1.2);
```

**Diagramas de Dispersión con plt.scatter**

* Un segundo método, más potente, para crear diagramas de dispersión es la función plt.scatter, que se puede utilizar de manera muy similar a la función plt.plot.

```python
fig = plt.figure()
ax = plt.axes()

ax.scatter(x,y,marker='o',s=50)
#Nos ayuda al ajuste automatico de los axis
ax.axis('tight');
```

* La diferencia principal de plt.scatter con respecto a plt.plot es que puede usarse para crear gráficos de dispersión donde las propiedades de cada punto individual (tamaño, color de relleno, color de borde, etc.) pueden ser controladas de manera individual o mapeadas a datos.

* Vamos a demostrar esto creando un gráfico de dispersión aleatorio con puntos de varios colores y tamaños. Para poder ver mejor los resultados superpuestos, también utilizaremos la palabra clave alpha para ajustar el nivel de transparencia.

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

fig = plt.figure()
ax = plt.axes()

sc = ax.scatter(x,y,c=colors, s=sizes, alpha=0.3, cmap='viridis')
fig.colorbar(sc,ax=ax);
```

* Observa que el argumento color se mapea automáticamente a una escala de colores (mostrada aquí mediante el comando colorbar()), y el argumento size se expresa en píxeles. De esta manera, el color y el tamaño de los puntos pueden utilizarse para transmitir información en la visualización, con el fin de ilustrar datos multidimensionales.

* Por ejemplo, podríamos usar los datos de Iris de Scikit-Learn, donde cada muestra es uno de tres tipos de flores cuyo tamaño de pétalos y sépalos ha sido medido cuidadosamente.





```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.2,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```

**Visualización de Errores**

* Para cualquier medición científica, la precisa contabilización de los errores es casi tan importante, si no más importante, que el reporte exacto del número en sí mismo. Por ejemplo, imagina que estoy utilizando algunas observaciones astrofísicas para estimar la Constante de Hubble, la medición local de la tasa de expansión del universo. Sé que la literatura actual sugiere un valor de alrededor de 71 (km/s)/Mpc, y mido un valor de 74 (km/s)/Mpc con mi método. ¿Son consistentes los valores? La única respuesta correcta, dado esta información, es la siguiente: no hay forma de saberlo.

* Supongamos que añado esta información con incertidumbres reportadas: la literatura actual sugiere un valor de alrededor de 71 ± 2.5 (km/s)/Mpc, y mi método ha medido un valor de 74 ± 5 (km/s)/Mpc. ¿Son consistentes los valores ahora? Esa es una pregunta que puede ser respondida cuantitativamente.

* En la visualización de datos y resultados, mostrar estos errores de manera efectiva puede hacer que un gráfico transmita información mucho más completa.


```python
print(plt.style.available)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

# Crear la figura y los ejes
# Usar plt.subplots() para crear la figura y los ejes al mismo tiempo
fig, ax = plt.subplots()  

# Aplicar un estilo
plt.style.use('seaborn-v0_8-whitegrid')

# Crear el gráfico de dispersión con barras de error
# fmt='.k' para puntos negros, yerr para barras de error verticales
ax.errorbar(x, y, yerr=dy, fmt='.k')  

# Configurar el título y etiquetas si es necesario
ax.set_title("Gráfico con barras de error")
ax.set_xlabel("Eje X")
ax.set_ylabel("Eje Y")

# Mostrar el gráfico
plt.show()

```

Además de estas opciones básicas, la función errorbar tiene muchas opciones para afinar las salidas. Utilizando estas opciones adicionales, puedes personalizar fácilmente la estética de tu gráfico de barras de error. A menudo encuentro útil, especialmente en gráficos con muchos elementos, hacer que las barras de error sean más claras que los propios puntos.





```python
plt.errorbar(x, y, yerr=dy, fmt='o', color='black',
             ecolor='lightgray', elinewidth=3, capsize=0);
```

**Errores Continuos**

* En algunas situaciones es deseable mostrar barras de error en cantidades continuas. Aunque Matplotlib no tiene una rutina incorporada para este tipo de aplicación, es relativamente fácil combinar primitivas como **plt.plot y plt.fill_between** para obtener un resultado útil.

* Aquí realizaremos una simple regresión de procesos gaussianos (GPR, por sus siglas en inglés), utilizando la API de Scikit-Learn (consulta "Introducción a Scikit-Learn" en la página 343 para más detalles). Este es un método para ajustar una función no paramétrica muy flexible a datos con una medida continua de incertidumbre. No profundizaremos en los detalles de la regresión de procesos gaussianos en este punto, sino que nos enfocaremos en cómo podrías visualizar una medición continua de error:

```python
import numpy as np
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF

# Define the model and draw some data
model = lambda x: x * np.sin(x)
xdata = np.array([1, 3, 5, 6, 8])
ydata = model(xdata)

# Define the kernel and Gaussian process regressor
kernel = RBF(length_scale=1.0)
gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=10)

# Fit the model
gp.fit(xdata[:, np.newaxis], ydata)

# Compute the Gaussian process fit
xfit = np.linspace(0, 10, 1000)
yfit, MSE = gp.predict(xfit[:, np.newaxis], return_std=True)
dyfit = 2 * MSE  # 2*sigma ~ 95% confidence region

# Plot the results (optional)
plt.plot(xdata, ydata, 'or')
plt.plot(xfit, yfit, '-', color='gray')
plt.fill_between(xfit, yfit - dyfit, yfit + dyfit,
color='gray', alpha=0.2)
plt.xlim(0, 10);
```
