**Filtro similar a Excel**

* Escogimos todos los jugadores que pertenecen al barcelona

```python
fifa_barcelona = df[df["Club"] == 'FC Barcelona']
print(type(fifa_barcelona))
print(fifa_barcelona.shape)
fifa_barcelona.head()
```

```python
fifa_no_barcelona = df[df["Club"] != "FC Barcelona"]
print(fifa_no_barcelona.shape)
fifa_no_barcelona.head()
```

```python
fifa_menor_30 = df[df["Age"] <= 30]
print(fifa_menor_30.shape)
fifa_menor_30.head()
```

**Filtro 2 categorias**

* jugadores mayores de 30 y brazileños

```python
fifa_mas_30_brazil = df[(df["Age"] > 30) & (df["Nationality"] == "Brazil")]
print(fifa_mas_30_brazil.shape)
fifa_mas_30_brazil.head()
```

* jugadores mayores de 30 o brazileños

```python
fifa_mas_30_o_brazil = df[(df["Age"] > 30) | (df["Nationality"] == "Brazil")]
print(fifa_mas_30_o_brazil.shape)
fifa_mas_30_o_brazil.head()
```

**Con str.contains** 
* podemos verificar si la columna contiene algun palabra especial para hacer el filtrado** 

```python
fifa_barc =df[df["Name"].str.contains("Sergio")]
print(fifa_barc.shape)
fifa_barc.head()
```

*Funcion: **isin***

```python
lista = ["Brazil","Portugal","Mexico"]
fifa_lista = df[df["Nationality"].isin(lista)]
print(fifa_lista.shape)
fifa_lista.head(20)
```
