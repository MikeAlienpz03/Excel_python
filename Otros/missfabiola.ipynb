```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(
    {
        "up" : np.arange(0,400,step=50).tolist(),
        "cft" : [800 for _ in range(8)],
        "cvt" : [0,100,240,360,460,520,540,750],
    }
)
df
```

```python
df["ct"] = df["cft"] + df["cvt"]
df
```

```python
df["cunit"] = df["ct"] / df["up"]
#redondeamos a 2 decimales
df["cunit"] = df["cunit"].round(2)
df
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(df["up"],df["cvt"],'--',color="r",alpha=0.6,label="cvt")
ax.plot(df["up"],df["ct"],'--',color="g",alpha=0.6,label="ct")
ax.plot(df["up"],df["cft"],color="b",alpha=1,label="cft")
ax.grid()
ax.axis('tight')
ax.legend(loc='best')
```
